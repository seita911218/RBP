# Risk parity

## Introduction

Risk parity is an investment strategy that focuses on allocating risk, rather than capital, equally across the various components of a portfolio. This approach differs significantly from traditional asset allocation strategies that allocate a fixed percentage of a portfolio's capital to various asset classes.

The primary goal of risk parity is to achieve a more stable and diversified portfolio. This approach can potentially offer better risk-adjusted returns, especially in volatile markets, as it avoids over-exposure to high-risk assets.


## Risk Allocation Principle

### Motivation

In Markowitz's Mean-Variance Portfolio, we solve the following problem.

$$
\begin{align*}
{\bf w}^* = &\text{arg } \underset{\bf w }{\text{max }} {\bf w}^T\mu - \lambda{\bf w}^T\Sigma{\bf w}
\\
\text{u.c. } &\sum^{n}_{i=1} w_i = 1
\end{align*}
$$

where ${\bf w} \in \mathbb{R}^{n}$ is the weight vector, $\mu \in \mathbb{R}^{n}$ is the mean of the portfoio, $\Sigma \in \text{Mat}_{n}(\mathbb{R})$ is the covariance matrix of the portfolio

Thus the optimization is to maximize the return and reduce risks within limits (depend on the trade-off parameter $\lambda$) in the meantime.

The issue with Markowitz's Portfolio Theory lies in its lack of control over specific risks. While the approach effectively reduces overall risk through optimization, it fails to address the diversification of risks. This oversight can lead to an excessive concentration of investments in high-performing assets.

Hence we want to study about how individual risks being allocated in a portfolio.

### Definitions

- **Risk measure**

    Let $\mathcal{R}({\bf w})$ be the risk measure of portfolio ${\bf w}$, there are several risk measures commonly used in finance, for instance: Volatility, VaR, CVaR.

    **Conherent property**
    
    Following Artzner et al. (1999), $\mathcal{R}({\bf w})$ is said to be **coherent**, if it satisfies the following properties:

    1. **Subadditivity**:  

        $\mathcal{R} ({\bf w}_1 + {\bf w}_2) \leq \mathcal{R} ({\bf w}_1) + \mathcal{R} ({\bf w}_2)$

        The risk of two portfolios should be less than the sum of the risks of the two separate portfolios.  

    2. **Homogeneity**:  

        $\mathcal{R} (\lambda {\bf w}) = \lambda \mathcal{R} ({\bf w})$ if $\lambda \geq 0$

        Leveraging or deleveraging the portfolio increases or decreases the risk measure in the same magnitude.  

    3. **Monotonicity**:  

        If ${\bf w}_1 \prec {\bf w}_2$, then $\mathcal{R} ({\bf w}_1) \geq \mathcal{R} ({\bf w}_2)$

        If portfolio ${\bf w}_2$ has a better return than portfolio ${\bf w}_1$ under all scenarios, risk measure $\mathcal{R}({\bf w}_1)$ should be higher than risk measure $\mathcal{R}({\bf w}_2)$.
        
    4. **Translation Invariance**:  

        If $m \in \mathbb{R}$, then $\mathcal{R} ({\bf w} + m) = \mathcal{R} ({\bf w}) - m$
        
        Adding a cash position of amount $m$ to the portfolio reduces the risk by $m$.

    **Convexity property**

    $\mathcal{R}({\bf w})$ is said to be **convex**, if it satisfies:

    $\mathcal{R}(\lambda {\bf w}_1 + (1-\lambda){\bf w}_2) \leq \lambda\mathcal{R}({\bf w}_1) + (1-\lambda)\mathcal{R}({\bf w}_2) \quad \forall \lambda \in [0,1]$

    **Remark**
    
    FÃ¶llmer and Schied (2002) propose to replace the homogeneity and subadditivity by this weaker condition, which means that diversification should not increase the risk.

- **Profit and Loss of the Portfolio**

    Let $\Pi$ be the total profit or loss of the portfolio, and $\Pi_{i}$ be the profit or loss of the i-th asset in the portfolio, hence we have the following:
    
    $$\Pi = \sum_{i=1}^n \Pi_{i}$$

- **Risk-Adjusted Performance Measure**

    The risk-adjusted performance measure $\text{RAPM}$ is defined as the expected return of the portfolio divided by the risk of the portfolio.

    which is a measure used to examine the performance of an investment portfolio by adjusting for the risk involved.

    $$
    \begin{align*}
    \text{RAPM}(\Pi) &:= \frac{\mathbb{E}[\Pi]}{\mathcal{R}(\Pi)}
    \\
    \text{RAPM}(\Pi_i|\Pi) &:= \frac{\mathbb{E}[\Pi_i]}{\mathcal{R}(\Pi_i|\Pi)}
    \end{align*}
    $$

    where
    
    $\mathcal{R}(\Pi)$ is the risk measure associated with profit and loss of portfolio

    $\mathcal{R}(\Pi_i|\Pi)$ is the risk contributions based on profit and loss to protfolio risk $\mathcal{R}(\Pi)$

### **Theorem I(Tasche)**

If the following conditions holds:

1.  Full allocation property:

    The total risk of the portfolio is the sum of the risks of each asset given the portfolio.

    $$
    \mathcal{R}(\Pi) = \sum_{i=1}^n \mathcal{R}(\Pi_i|\Pi)
    $$
    
2. $\text{RAPM}$ compatibility:  

    There exists a positive $\epsilon_i$ such that if the risk-adjusted performance measure of an asset given the portfolio is greater than the risk-adjusted performance measure of the portfolio,  
    then increasing the weight of the asset in the portfolio will increase the risk-adjusted performance measure of the portfolio.
    
    $$
    \begin{align*}
    \exists \epsilon_{i} > 0
    \quad
    \text{such that}
    \quad
    \text{RAPM}(\Pi_i|\Pi) > \text{RAPM}(\Pi) \Rightarrow \text{RAPM}(\Pi + h\Pi_i) > \text{RAPM}(\Pi) \quad \forall \ 0 < h < \epsilon_i
    \end{align*}
    $$

Then the risk contribution of an individual asset given the portfolio $\mathcal{R}(\Pi_i|\Pi)$ can be calculated as follows:

$$
\mathcal{R}(\Pi_i|\Pi)=\frac{d}{dh}\mathcal{R}(h\Pi_i+\Pi)|_{h=0}
$$

### Remark

In case of a subadditive risk measure, we have: $\mathcal{R}(\Pi_i|\Pi) \leq \mathcal{R}(\Pi_i)$

This means that the risk contributions of asset i is always than its stand-along risk measure. The difference is related to the risk diverification.

### Definitions

- **Risk Contribution**

    The risk contribution of asset $\text{i}$ in terms of weight $\mathcal{RC}_{i}({\bf w})$ is the contribution of an individual asset to the overall risk of the portfolio, by **Theroem I** it can be uniquely defined as:

    $$
    \mathcal{RC}_{i}({\bf w}) = w_{i} \frac{\partial \mathcal{R}(\bf w)}{\partial w_{i}}
    $$


### **Theorem II (Euler's Theorem)**

Let ${\bf w} \in \mathbb{R}^{n}$, and $f:\mathbb{R}^{n} \longrightarrow \mathbb{R}$ is a continuously differentiable function and a positively homogeneous function of degree one, which means:

$$
f(c{\bf w})=cf({\bf w}) \quad \forall c>0
$$

Then $f$ can be expressed as:

$$
f({\bf w})=\sum_{i=1}^n w_i\frac{\partial f}{\partial w_i}
$$

**Proof**

Let $g(t)$ be a function defined as follows:

$$
g(t)=f(c{\bf w}) \tag{1}
$$

Since $f$ a positively homogeneous function of degree one, which means $g(t)$ also equals to:

$$
g(t)=cf({\bf w}) \tag{2}
$$

Consider the derivative of $g(t)$ with respect to $t$:

By $\text{(1)}$:

$$
g'(t) = \frac{\mathrm{d}}{\mathrm{d}t}f(t\textbf{w}) = \sum_{i=1}^n \frac{\partial f(t{\bf w})}{\partial (t w_i)} \cdot \frac{\partial (tw_i)}{\partial t} = \sum_{i=1}^n \frac{\partial f(t{\bf w})}{\partial (tw_i)} w_i
$$

By $\text{(2)}$:

$$
g'(t) = \frac{\mathrm{d}}{\mathrm{d}t}t f({\bf w}) = f({\bf w})
$$

Hence we have:

$$
g'(t) = f({\bf w}) = \sum_{i=1}^n  w_i \frac{\partial f(t{\bf w})}{\partial (tw_i)}
$$

In particular, choose $t=1$, we get the desired result:

$$
f({\bf w}) = \sum_{i=1}^n w_i \frac{\partial f({\bf w})}{\partial (w_i)} \quad \Box
$$

### **Application(Euler allocation principle)**

If the risk measure $\mathcal{R}({\bf w})$ satisfies **Homogeneity**, then the total risk of the portfolio is the sum of all risk contribution of each assets:

$$
\mathcal{R}({\bf w}) = \sum_{i=1}^n \mathcal{RC}_i
$$

**Proof**

Directy apply Euler's Theorem, then we get the desired result:

$$
\mathcal{R}({\bf w}) = \sum_{i=1}^n w_i \frac{\partial \mathcal{R}({\bf w})}{\partial (w_i)} = \sum_{i=1}^n \mathcal{RC}_i \quad \Box
$$
        
### Remark

We can always define the risk contributions of a risk measure by $\mathcal{RC}_{i}({\bf w}) = w_{i} \frac{\partial \mathcal{R}(\bf w)}{\partial w_{i}} $, however this does not mean that it satisfies Euler allocation principle.

## Risk Budgeting Portfolio

**Defnition**

A budget vector ${\bf b}=(b_1, \cdots , b_n)^T \in \mathbb{R}^{n}$, where $b_{i} \geq 0 \quad \forall i$ and $\sum^{n}_{i=1} b_{i}=1$, is a predetermined percentage risk contribution target for all the assets, it satisfies:

$$
\mathcal{R}({\bf w}) \cdot {\bf b} = \left(\mathcal{RC}_1({\bf w}), \cdots, \mathcal{RC}_n({\bf w}) \right)
$$

**Definition (Relative Risk Contribution)**

Relative risk contribution ($\mathcal{RRC}$) is defined as:

$$
\mathcal{RRC}_i({\bf w})= \frac{\mathcal{RC}_i({\bf w})}{\mathcal{R}({\bf w})}
$$

In the Risk Budgeting Portfolio, we have:

$$
\mathcal{RRC}_i({\bf w})= b_i \quad \forall i
$$

```python
class Stocks():
  def __init__(self, arg_stocks_list=[], arg_begin='2015-01-01', arg_end='2023-10-01'):
    """
    Initialize the Stocks class.

    Parameters:
    - arg_stocks_list: List of stock symbols.
      For example:
        ['2330.TW','2337.TW','2357.TW','2454.TW','3231.TW','3443.TW'] or
        ['TSM','AAPL','GOOG','MSFT','AMZN','GOOGL']
    - arg_begin: Start date for the stock data.
    - arg_end: End date for the stock data.

    Returns:
    - None: This function doesn't return anything but initializes the class instance.
    """
    self.stocks_list = arg_stocks_list  # List of stock symbols
    self.stocks_list.sort()

    self.start_date = arg_begin  # Start date
    self.end_date = arg_end  # End date

    self.N = len(self.stocks_list)  # Number of stocks

    self.data = None
    self.set_data()

    self.M = len(self.data)  # The number of rows in the data

    self.rtns = None
    self.set_rtns()

    self.expected_rtns = None
    self.correlation_matrix = None
    self.std_dev = None
    self.covariance_matrix = None
    self.set_basic_elements(self.rtns)

  def set_data(self):
    """
    This function downloads stock data for a list of stocks and stores it in a DataFrame.

    Parameters:
    - self: The instance of the class where the function is called.

    Returns:
    - None: The function modifies the 'data' attribute of the 'self' object, it does not return anything.
    """
    self.data = pd.DataFrame()  # Initialize an empty DataFrame to store data
    # Iterate over each stock in the list of stocks
    for stock in self.stocks_list:
      # Download the stock data using the yfinance library
      df = yf.download(stock, start=self.start_date, end=self.end_date)
      # Select only the adjusted closing price from the downloaded data
      df = df[['Adj Close']]
      # Rename the column to the stock symbol for clarity
      df = df.rename(columns = {'Adj Close': stock})
      # Check if the main DataFrame is empty
      if self.data.empty:
        # If it is, assign the downloaded data to the main DataFrame
        self.data = df
      else:
        # If it's not, concatenate the downloaded data to the main DataFrame
        self.data = pd.concat([self.data, df], axis=1)
      # Print the stock symbol to indicate the progress
      print(stock)
    # The function does not return anything as it modifies the 'data' attribute in-place
    return

  def set_rtns(self):
    """
    Calculate the DataFrame of daily returns for each stock.

    Parameters:
    - self: the instance of the class.

    Returns:
    - None: this function doesn't return anything but updates the rtns attribute of the class instance.
    """
    self.rtns = pd.DataFrame()
    for stock in self.stocks_list:
      self.rtns[stock] = self.data[stock].pct_change() + 1
    self.rtns.fillna(1, inplace=True)
    return

  def rtns_cumprod(self):
    """
    Plot the cumulative product of returns.

    Parameters:
    - self: the instance of the class.

    Returns:
    - None: this function doesn't return anything but plots the rtns_cumprod attribute of the class.
    """
    self.rtns.cumprod(axis = 0).plot()
    return

  def set_expected_rtns(self,arg_rtns):
    """
    Calculate the expected returns.

    Parameters:
    - self: the instance of the class.
    - arg_rtns: DataFrame of returns.

    Returns:
    - None: this function doesn't return anything but updates the expected_rtns attribute of the class instance.
    """
    self.expected_rtns = arg_rtns.mean()
    return
  def set_correlation(self,arg_rtns):
    """
    Calculate the correlation matrix.

    Parameters:
    - self: the instance of the class.
    - arg_rtns: DataFrame of returns.

    Returns:
    - None: this function doesn't return anything but updates the correlation_matrix attribute of the class instance.
    """
    self.correlation_matrix = np.corrcoef(arg_rtns,None,False)
    return
  def set_std_dev(self,arg_rtns):
    """
    Calculate the standard deviation.

    Parameters:
    - self: the instance of the class.
    - arg_rtns: DataFrame of returns.

    Returns:
    - None: this function doesn't return anything but updates the std_dev attribute of the class instance.
    """
    self.std_dev = np.std(arg_rtns)
    return
  def set_cov(self):
    """
    Calculate the covariance matrix.

    Parameters:
    - self: the instance of the class.

    Returns:
    - None: this function doesn't return anything but updates the covariance_matrix attribute of the class instance.
    """
    self.covariance_matrix = np.zeros([self.N,self.N])
    for i in range(self.N):
      for j in range(self.N):
        self.covariance_matrix[i][j]=self.correlation_matrix[i][j]*self.std_dev[i]*self.std_dev[j]
    return
  def set_basic_elements(self,arg_rtns):
    """
    Calculate the basic elements.

    Parameters:
    - self: the instance of the class.
    - arg_rtns: DataFrame of returns.

    Returns:
    - None: this function doesn't return anything but updates the expected_rtns, correlation_matrix, std_dev, and covariance_matrix attributes of the class instance.
    """
    self.set_expected_rtns(arg_rtns)
    self.set_correlation(arg_rtns)
    self.set_std_dev(arg_rtns)
    self.set_cov()
    return

#-----------------------------------------------------------------------------------------------------------------------------------------------------------------

# Get return
def get_return(stock_arr, start, end):
    """
    Retrieve the daily returns of stocks.

    Parameters:
    - stock_arr: List of stock tickers.
    - start: Start date for data retrieval.
    - end: End date for data retrieval.

    Returns:
    - returns: DataFrame of daily returns for each stock.
    """
    n = len(stock_arr)
    stocks = yf.download(stock_arr, start=start, end=end)

    # Calculate the daily returns
    returns = stocks['Adj Close'].pct_change()
    returns = returns.fillna(0)

    return returns

# Get sharp_ratio
def sharp_ratio(stoke_return,risk_free=0.043):
  """
    Calculate the sharp ratio.

    Parameters:
    - stoke_return: DataFrame of stock return.
    - risk_free: risk_free rate

    Returns:
    - returns: DataFrame of daily returns for each stock.
    """
  pct_return=stoke_return.pct_change()
  return (pct_return.mean()-risk_free/365)/pct_return.std()

def print_SR_sr(arg_result):
  # arg_result=backtesting_start_end(day_equal,weight_equal,n)
  # pct_return=arg_result.loc[:,"strategy_cum_return"]
  sr=empyrical.sharpe_ratio(arg_result.iloc[:,-1].pct_change())
  SR=sharp_ratio(arg_result.iloc[:,-1],0)
  print(SR*252**0.5,sr)
  # print(arg_result.iloc[:,-1].pct_change())
  return

def list_cov(returns_train, n, interval=7, cov_days=30):
    """
    Computes the covariance matrix for the training data.

    Parameters:
    - returns_train: DataFrame, historical returns of stocks for training.
    - n: int, number of stocks.
    - interval: int, interval for selecting data.
    - cov_days: int, number of days for covariance calculation.

    Returns:
    List of covariance matrices.
    """
    cov = []
    for i in range(len(returns_train) - cov_days - 1, 1, -interval):
        cov.append((returns_train.iloc[-1 - i - cov_days:-1 - i, 0:n]).cov())
    return cov

def list_mu(returns_train, n, interval=7, cov_days=30):
    """
    Computes the covariance matrix for the training data.

    Parameters:
    - returns_train: DataFrame, historical returns of stocks for training.
    - n: int, number of stocks.
    - interval: int, interval for selecting data.
    - cov_days: int, number of days for covariance calculation.

    Returns:
    List of covariance matrices.
    """
    mu = []
    for i in range(len(returns_train) - cov_days - 1, 1, -interval):
        mu.append((returns_train.iloc[-1 - i - cov_days:-1 - i, 0:n]).mean(axis=0))
    return mu
```

## Main Problem

In Risk Budgeting Portfolio, given the budget vector ${\bf b}=(b_1, \cdots ,b_n)^T$, we consider the following **Non-linear System**:

$$
\begin{align*}
\mathcal{RC}_i({\bf w}) = b_i \mathcal{R}({\bf w}) \tag{3}
\\
\text{u.c. }
\begin{cases}
w_i \in [0,1] \quad \forall i
\\
\sum^n_{i=1} w_i =1
\end{cases}
\end{align*}
$$

which means we want to find the weight ${\bf w} \in \mathbb{R}^n$ vector such that the risk contribution of the portfolio is equals to the value ${\bf b} \in \mathbb{R}^n$ we seted.

### Analysis to Main Problem (Existance and Uniqueness)

In [R1] author says that there are some problems solving **Non-linear System** (3):

1. In general, it is not possible to find a analytical solution.

2. We can always find the numerical solution, but they do not always converge.

This is why it is better to transform the **Non-linear System** into an **Optimization Problem**(Bruder, Roncalli):

$$
\begin{align*}
&{\bf w}^* = \text{arg } \underset{{\bf w}}{\text{min}} \text{ } \mathcal{R}({\bf w}) \tag{4}
\\
\text{u.c. }
&\begin{cases}
\sum^n_{i=1} b_i \log w_i \geq c
\\
w_i \geq 0 \quad \forall i
\end{cases}
\end{align*}
$$

where $c$ is an arbitrary constant.

### Theorem

The solution to the optimization problem depends on the values of the budget vector ${\bf b}$ .

1. If $b_i > 0$ for all $i$, then a solution exists and is unique.
2. If $b_i \ge 0$ for all $i$, then there may be several solutions.

**Proof(for the case ${\bf b}$ > 0)**

We use the Lagrange method, with Karush-Kuhn-Tucker ([KKT](https://en.wikipedia.org/wiki/Karush%E2%80%93Kuhn%E2%80%93Tucker_conditions)) conditions to solve this optimization problem.

First, we define a Lagrange function $\mathcal{L}({\bf y},{\bf \lambda},\lambda_c)$:

$$
\mathcal{L}({\bf y},{\bf \lambda},\lambda_c) = \mathcal{R}({\bf y}) - {\bf \lambda}^T{\bf w}- \lambda_c(\sum_{i=1}^n b_i\log{y_i}-c)
$$

with ${\bf y}\text{, } {\bf \lambda} \in \mathbb{R}^{n}$, and $\lambda_c \in \mathbb{R}$

Next, we set the gradient of function $\mathcal{L}({\bf y},{\bf \lambda},\lambda_c)$ to zero, which gives us:

$$
\frac{\partial}{\partial y_i} \mathcal{L} ({\bf y},{\bf \lambda},\lambda_c) =  \frac{\partial}{\partial y_i} \mathcal{R}({\bf y}) - \lambda_i - \lambda_c\frac{b_i}{y_i} = 0 \tag{5}
$$

While the KKT conditions are:

$$
\begin{cases}
\text{1. min}\{\lambda_i,y_i\} = 0
\\
\text{2. min}\{\lambda_c,\sum^n_{i=1} b_i \log y_i -c\} = 0
\end{cases}
$$

Because $\log y_i$ is not defined for $y_i = 0$, it follows that $y_i > 0$, hence by KKT condition 1 we have: $\lambda_i=0$.

Furthermore since ${\bf y} \neq 0$, hence $\sum^n_{i=1} b_i \log y_i = c$ is necessarily reached, hence by KKT condition 2 we have: $\lambda_c > 0$

Hence by equation (5):

$$
y_i\frac{\partial \mathcal{R}({\bf y})}{\partial y_i} = \lambda_c b_i
$$

which means:

$$
\mathcal{RC}_i({\bf y}) = \lambda_c b_i
$$

Hence implies:

$$
\frac{\mathcal{RC}_i({\bf y})}{\mathcal{RC}_j({\bf y})} = \frac{b_i}{b_j}
$$

This condition gave us the unique optimal solution ${\bf y}^*$.

Finally, we can obtain the solution ${\bf w}^*$ of (4) by normalizing ${\bf y}^*$.

$$
w^*_i = \frac{y^*_i}{\sum^n_{j=1} y^*_j}
$$

### Risk budgeting versus weight budgeting

1. **Risk Budgeting (RB)**:  

    In RB portfolio, we find the weight vector ${\bf w} \in \mathbb{R}^n$ such that the risk contribution match the budget vector ${\bf b} \in \mathbb{R}^n$:

    $$
    \mathcal{RC}_i({\bf w}) = b_i\mathcal{R}({\bf w})
    $$

    and so we have the following relationship:
    
    $$
    \frac{\mathcal{RC}_i}{b_i}=\frac{\mathcal{RC}_j}{b_j} \quad \forall i,j
    $$


2. **Weight Budgeting (WB)**:  

    In WB portfolio, we assign to each assets a weight that is equal to its budget:

    $$
    {\bf w} = {\bf b}
    $$
    
    and so we have the following relationship:

    $$
    \frac{w_i}{b_i}=\frac{w_j}{b_j} \quad \forall i,j
    $$

3. **Minimum Risk (MR)**:  

    In MR portfolio, we solve the optimization problem:

    $$
    \begin{align*}
    &{\bf w}^* = \text{arg } \underset{{\bf w}}{\text{min }} \mathcal{R}({\bf w})
    \\
    \text{u.c. } &\sum^n_{i=1} w_i=1
    \end{align*}
    $$

    and so we have the following relationship:

    $$
    \frac{\partial\mathcal{R}({\bf w})}{\partial w_i}=\frac{\partial\mathcal{R}({\bf w})}{\partial w_j} \quad \forall i,j
    $$

### Remark

If in particular choose uniform budget ${\bf b}=(\frac{1}{n},\cdots,\frac{1}{n})^T$:

1. **Risk Budgeting (RB) $\rightarrow$ Equal Risk Contribution (ERC)**:  

    In ERC portfolio, we find the weight vector ${\bf w} \in \mathbb{R}^n$ such that the risk contribution match the budget vector ${\bf b} \in \mathbb{R}^n$:

    $$
    \mathcal{RC}_i({\bf w}) = \frac{\mathcal{R}({\bf w})}{n}
    $$


2. **Weight Budgeting (WB) $\rightarrow$ Equally Weight (EW)**:  

    In EW portfolio, we assign to each assets a weight that is equally weighted:

    $$
    {\bf w} = \frac{1}{n}
    $$
    
### Risk Comparison

We have the following relationship between total risk of RB, WB, MR portfolios.

$$\mathcal{R}_{MR} \leq \mathcal{R}_{RB} \leq \mathcal{R}_{WB}$$

**Proof:**

Consider the previous optimization:

$$
\begin{align*}
&{\bf w}^* = \text{arg } \underset{{\bf w}}{\text{min}} \text{ } \mathcal{R}({\bf w}) \tag{4}
\\
\text{u.c. }
&\begin{cases}
\sum^n_{i=1} b_i \log w_i \geq c
\\
w_i \geq 0 \quad \forall i
\end{cases}
\end{align*}
$$

Then these three cases can be viewed as optimization with different $c$.

- $\mathcal{R}_{MR}: c = -\infty$, since it vanishes the first constrain of (4)

- $\mathcal{R}_{RB}:$ choose $c^* = c - \log(\sum^n_{i=1} y^*_i)$ in (4) $\in (-\infty,\sum^n_{i=1} b_i\log{b_i}]$

- $\mathcal{R}_{WB}: c = \sum_{i=1}^n b_i\ln b_i$, since it implies $\frac{w_i}{b_i}=\frac{w_j}{b_j} \quad \forall i,j$

and we have the following property:

$$
\text{If } c_1 \leq c_2 \quad \text{then } \mathcal{R}\left({\bf w}^*(c_1)\right) \leq \mathcal{R}\left({\bf w}^*(c_2)\right)
$$

Thus, for $c \in [-\infty,\sum^n_{i=1} b_i\log{b_i}]$ we have:

$$
\mathcal{R}\left({\bf w}^*(-\infty)\right) \leq \mathcal{R}\left({\bf w}^*(c)\right) \leq \mathcal{R}\left({\bf w}^*(\sum^n_{i=1} b_i\log{b_i})\right)
$$

In other words, we have:

$$
\mathcal{R}_{MR} \leq \mathcal{R}\left({\bf w}^*(c)\right) \leq \mathcal{R}_{WB}
$$

In particular, choose $c = c^*$ defined above we have:

$$
\mathcal{R}_{MR} \leq \mathcal{R}_{RB} \leq \mathcal{R}_{WB} \quad \Box
$$

### Remark

In particular, if we choose risk measure as volitility risk measure $\mathcal{R}({\bf w})=\sigma({\bf w})$ and uniform budget ${\bf b}=(\frac{1}{n},\cdots,\frac{1}{n})$, we have:

$$
\sigma_{MV} \leq \sigma_{ERC} \leq \sigma_{EW}
$$


## Risk Measure: Volatility(Standard Deviation)

In portfolio theory, we often use volatility(standard deviation) as measures of risk.  

$$\mathcal{R}({\bf w}) = \sigma({\bf w})$$

### Definition

- **Variance**

    The variance of a portfolio $\sigma^2({\bf w})$ with weight ${\bf w} \in \mathbb{R}^{n}$ with covariance matrix $\Sigma \in \text{Mat}_n(\mathbb{R})$ is calculated as:

    $$
    \sigma^2({\bf w}) = {\bf w}^T\Sigma{\bf w}
    $$

- **Standard Deviation**

    The standard deviation of a portfolio $\sigma({\bf w})$ with weight ${\bf w} \in \mathbb{R}^{n}$ with covariance matrix $\Sigma \in \text{Mat}_n(\mathbb{R})$ is calculated as:

    $$
    \sigma({\bf w}) = \sqrt{{\bf w}^T\Sigma{\bf w}}
    $$

### Lemma

For $c>0$

Standard deviation is a positively homogeneous function of degree one:

$$
\begin{align*}
\sigma(c{\bf w}) &= \sqrt{(c{\bf w})^T \Sigma (c{\bf w})}
\\
&= \sqrt{c^2 {\bf w}^T\Sigma{\bf w}}
\\
&= \left|c\right| \sigma({\bf w})
\\
&= c \sigma({\bf w})
\end{align*}
$$

Apply Euler's Theorem, we obtain:

$$
\sigma({\bf w}) = \sum^n_{i=1} w_i\frac{\partial \sigma({\bf w})}{\partial w_i}
$$

Therefore, we often take volatility (standard deviation) as the risk measure in portfolio theory:

$$
\mathcal{R}({\bf w}) = \sigma({\bf w}) = \sqrt{{\bf w}^T\Sigma {\bf w}}
$$

Hence we have:

- **Risk Contribution**

$$
\mathcal{RC}_i({\bf w})= w_i\frac{\partial \mathcal{R}({\bf w})}{\partial w_i} = \frac{w_i(\Sigma {\bf w})_i}{\sqrt{{\bf w}^T\Sigma {\bf w}}}
$$

- **Relative Risk Contribution**

$$
\mathcal{RRC}_i({\bf w})= \frac{\mathcal{RC}_i({\bf w})}{\mathcal{R}({\bf w})} = \frac{w_i(\Sigma {\bf w})_i}{{\bf w}^T\Sigma {\bf w}}
$$


## Back to Main Problem

We now use the volatility as risk measure to solve the previous **Non-linear System** (3):

$$
\begin{align*}
\mathcal{RC}_i({\bf w}) = b_i \mathcal{R}({\bf w})
\\
\text{u.c. }
\begin{cases}
w_i \in [0,1] \quad \forall i
\\
\sum^n_{i=1} w_i =1
\end{cases}
\end{align*}
$$

which is:

$$
\begin{align*}
w_i(\Sigma {\bf w})_i = b_i{\bf w}^T\Sigma {\bf w}
\\
\text{u.c. }
\begin{cases}
w_i \in [0,1] \quad \forall i
\\
\sum^n_{i=1} w_i =1
\end{cases}
\end{align*}
$$

### Convex Optimization Solution

To slove the equation (6):

We first introduce a Convex Optimization Problem, and explain why they works, and give 2 algorithm for the solution of the Convex Optimization Problem.

**Definition**

A function is called **convex** if

$$
f(\lambda x_1+(1-\lambda )x_2) \le \lambda f(x_1)+(1-\lambda )f(x_2) \quad \forall 0 \leq \lambda \leq 1
$$

**Remark**

The following Convex Optimization Problem have some issues:

- This only solves the simplest situation with a simplex constraint $w_i \geq 0 \quad \forall i$ and $\sum^n_{i=1} w_i =1$ (long only)

- It **cannot** maximize the return ${\bf w}^T\mu$ or minimize the volatility ${\bf w}^T\Sigma {\bf w}$.

- It is **not always convex** in the general case.

**Spinu(2013) Convex Optimization**

Consider the following optimization problem:

Let $f({\bf x}) = \frac{1}{2}{\bf x}^T\Sigma{\bf x} - {\bf b}^T\log({\bf x})$ be a convex function

$$
\begin{align*}
{\bf x}^* &= \text{arg } \underset{{\bf x}}{\text{min }} \frac{1}{2}{\bf x}^T\Sigma{\bf x} - {\bf b}^T\log({\bf x})
\\
&\text{u.c. } x_i\geq 0 \quad \forall i
\end{align*}
$$

Then the gradient of $f({\bf x})$ is given by $\nabla f({\bf x}) = \Sigma {\bf x} -\frac{{\bf b}}{{\bf x}}$.

Setting $\nabla f({\bf x})=0$, we obtain the optimality condition $\Sigma {\bf x} = \frac{{\bf b}}{{\bf x}}$, which is we desired.

The normalization is ${\bf w} = \frac{{\bf x}}{\sum^n_{i=1} x_i}$

#### Solution Algorithm

There are two primary methods to solve these convex optimization problems:

**1. The Newton Algorithm**

Given a Convex Optimization Problem:

$$
\begin{align*}
{\bf x}^* &= \text{arg } \underset{{\bf x}}{\text{min }} f({\bf x}) \quad \left(\text{or } {\bf x}^* = \text{arg } \underset{{\bf x}}{\text{max }} f({\bf x})\right)
\\
&\text{u.c. } x_i\geq 0 \quad \forall i
\end{align*}
$$

**Alogorithm**

The basic concept for Newton's algorithm is Tylor expansion.

- **Step 1**

    Choose an initial vector ${\bf x}^{(1)} \in \mathbb{R}^n$

- **Step 2**

    Use the following formula to update the vector ${\bf x}$

    $$
    {\bf x}^{(k+1)} = {\bf x}^{(k)} - H^{-1}({\bf x}^{(k)})\nabla f({\bf x}^{(k)}) \quad \left(\text{or } {\bf x}^{(k+1)} = {\bf x}^{(k)} + H^{-1}({\bf x}^{(k)})\nabla f({\bf x}^{(k)})\right)
    $$

    where
    
    $$
    H({\bf x}) =
    \begin{pmatrix}
    \frac{\partial^2 f}{\partial x_1^2} & \frac{\partial^2 f}{\partial x_1 \partial x_2} & \cdots & \frac{\partial^2 f}{\partial x_1 \partial x_n} \\
    \frac{\partial^2 f}{\partial x_2 \partial x_1} & \frac{\partial^2 f}{\partial x_2^2} & \cdots & \frac{\partial^2 f}{\partial x_2 \partial x_n} \\
    \vdots & \vdots & \ddots & \vdots \\
    \frac{\partial^2 f}{\partial x_n \partial x_1} & \frac{\partial^2 f}{\partial x_n \partial x_2} & \cdots & \frac{\partial^2 f}{\partial x_n^2} \\
    \end{pmatrix}
    $$
    
    is the Hessian matrix of $f$

- **Step 3**

    Repeat step 2, until the solution ${\bf x}$ converge, or it's precise enough.

    
**Adventage**

1. **Convergence Speed**

    Newton's algorithm, when applicable, typically converges to a solution faster than many other optimization methods, especially in well-behaved scenarios where the objective function is smooth and convex.

2. **Precision**

    It can provide highly accurate solutions by utilizing the second-order information of the objective function, making it powerful for applications where precision is critical.

**Disadventage**

1. **Computational Complexity**

    Newton's method requires the computation of the Hessian matrix (the second-order partial derivatives of the objective function) and its inversion, which can be computationally expensive and impractical for large-scale problems.

2. **Requirement for Derivatives**

    The necessity to compute derivatives can be a limitation in problems where the objective function's derivatives are difficult to calculate or are not available analytically.

**2. The Cyclical Coordinate Descent Algorithm (CCD)**

Given a Convex Optimization Problem:

$$
\begin{align*}
{\bf x}^* &= \text{arg } \underset{{\bf x}}{\text{min }} f({\bf x}) \quad \left(\text{or } {\bf x}^* = \text{arg } \underset{{\bf x}}{\text{max }} f({\bf x})\right)
\\
&\text{u.c. } x_i\geq 0 \quad \forall i
\end{align*}
$$

**Alogorithm**

- **Step 1**

    Choose an initial vector ${\bf x}^{(1)} \in \mathbb{R}^n$

- **Step 2**

    Use the following formula to update the vector ${\bf x}$

    $$
    x_i^{(k+1)} = \text{arg } \underset{x_i}{\text{min }} f(x_1^{(k)}, \cdots, x_{i-1}^{(k)}, x_i, x_{i+1}^{(k)}, \cdots, x_n^{(k)}) \quad \forall i \quad \left(\text{or } x_i^{(k+1)} = \text{arg } \underset{x_i}{\text{max }} f(x_1^{(k)}, \cdots, x_{i-1}^{(k)}, x_i, x_{i+1}^{(k)}, \cdots, x_n^{(k)}) \quad \forall i\right)
    $$

- **Step 3**

    Repeat step 2, until the solution ${\bf x}$ converge, or it's precise enough.

**Adventage**

1. Simplicity and Efficiency

    CCD is simpler to implement and can be more computationally efficient for large-scale problems because it optimizes one variable at a time while keeping others fixed, which often leads to a reduction in computational overhead.

2. Flexibility

    It can be applied to problems where the objective function or constraints are not differentiable, making it suitable for a broader range of applications.

**Disadventage**

1. Convergence Speed

    While simpler, CCD may require more iterations to converge to an optimal solution compared to methods like Newton's, especially in cases where the objective function landscape is complex.

2. Tuning Required

    The performance of CCD can be sensitive to the order in which coordinates are updated, and sometimes heuristic or problem-specific knowledge is required to choose the most effective update sequence.

```python
# Newton's Algorithm
def newton(w0, b, cov, max_err, max_iter=1000):
  """
  This function implements Newton's method to solve a convex optimization problem.

  Parameters:
  - w0: shape(n, 1) The initial guess for the optimization.
  - b: shape(n, 1) The vector of coefficients.
  - cov: shape(n, n) The covariance matrix.
  - max_err: The maximum error allowed for convergence.
  - max_iter: The maximum number of iterations.

  Returns:
  - w: array of Optimized vector of shape(n,1).
  """
  n = len(cov)

  # Define the objective function, its gradient, and Hessian
  f = lambda w: 0.5 * w.T.dot(cov.dot(w)) - b.T.dot(np.log(w))  # Objective function
  df = lambda w: cov.dot(w) - b / w  # Gradient of the objective function
  d2f = lambda w: cov + np.diag((b / (w ** 2)).reshape(n))  # Hessian of the objective function

  w = [w0]  # Initialize the solution
  err = 100  # Initialize the error
  i, t = 0, 0  # Initialize the iteration counters

  # Iterative optimization using Newton's method
  while err >= max_err and t < max_iter:
    if abs(w[-1][0]) > 1e4:
      # To prevent divergence, return an invalid result if w values become too large
      w.append([-1])
      return w

      # Update the solution using Newton's method
    w.append(w[i] - np.linalg.inv(d2f(w[i])).dot(df(w[i])))
    i += 1
    err = np.linalg.norm(w[i] - w[i - 1], ord=1)  # Compute the error
    t += 1

    # Ensure non-negativity of w values
    w[-1][w[-1] < 0] = 0.0001

  # Normalize the result to ensure it is a probability vector
  w = np.array(w).reshape(len(w), n)
  w = (w.T / w.sum(axis=1)).T

  return w
```

```python
# Cyclical Coordinate Descent Algorithm (CCD)
def CCD(w0, b, cov, max_err, max_iter=1000):
  """
  Implements the Cyclical Coordinate Descent (CCD) algorithm for solving a convex optimization problem.

  Parameters:
  - w0: ndarray of shape(n, 1). Initial guess for the optimization.
  - b: ndarray of shape(n, 1). Vector of coefficients.
  - cov: ndarray of shape(n, n). Covariance matrix.
  - max_err: float. Maximum error allowed for convergence.
  - max_iter: int. Maximum number of iterations.

  Returns:
  - w: ndarray of shape(n,1). Optimized vector.
  """
  n = len(cov)
  w = [w0]
  err = 100
  k, t = 0, 0

  # Iterative optimization using CCD
  while err >= max_err and t < max_iter:
    next_w = []

    # Update each coordinate separately
    for i in range(len(w0)):
      w_i = np.array(w[k])
      w_i[i] = 0
      next_w.append(
          (-(w_i.T @ cov[:, i]) + np.sqrt((w_i.T @ cov[:, i])**2 + 4 * cov[i, i] * b[i])) /
          (2 * cov[i, i])
          )

    # Append the updated vector to the result
    w.append(np.array(next_w).reshape(n, 1))

    # To prevent divergence, return an invalid result if w values become too large or too small
    if abs(w[-1][0]) > 1e4 or abs(w[-1][0]) < 1e-5:
      w.append([-1])
      return w

    k += 1
    err = np.abs(np.sum(w[k] - w[k - 1]))
    t += 1

  # Normalize the result to ensure it is a probability vector
  w = np.array(w).reshape(len(w), n)
  w = (w.T / np.sum(w, axis=1)).T

  return w
```

### General Non-convex Optimization Problem Solution

The previous methods are based on a convex reformulation of the problem so they are guaranteed to converge to the optimal risk budgeting solution, however, they can only be employed for the simplest risk budgeting formulation with a simplex constraint set.

In those more general cases, we need more sophisticated formulations, which unfortunately are not convex, We will overview the different general formulations and the solution methods.

In the following we will introduce a more general risk parity formulation.

**Definition**

A function $f$ is **strongly convex** if

$$
\exists \alpha > 0 \quad \text{such that} \quad f(x) - Î±||x||^2 \quad \text{is convex}
$$

A function $f$ is **uniformly strongly convex** if

$$
\exists m>0 \quad \text{such that} \quad f(x) - m||x||^2 \quad \text{is convex} \quad \forall x
$$

A function $f$ is **Lipschitz continuous** if

$$
\exists C \in \mathbb{R} \quad  \text{such that} \quad  |f(y) - f(x)| \leq C|y - x| \quad \forall x,y
$$

**Feng and Palomar (2015)**

Let $U({\bf w}) = \sum^n_{i=1} g_i^2({\bf w}) + \lambda F({\bf w})$

$$
\begin{align*}
{\bf w}^* &= \text{arg } \underset{{\bf w}}{\text{min }} U({\bf w})
\\
&\text{u.c.}
\begin{cases}
\sum^n_{i=1} w_i = 1
\\
\text{other convex constraints}
\end{cases}
\end{align*}
$$

where

$g_i({\bf w})$: risk concentration measurement, where $\mathcal{R}({\bf w})=\sum ^n_{i=1} g_i^2({\bf w})$, e.g. $g_i({\bf w}) = \frac{w_i(\Sigma{\bf w})_i}{{\bf w}^T\Sigma{\bf w}}-\frac{1}{n}$

$F({\bf w})$: preference, e.g. $F({\bf w}) = 0$ or $-{\bf \mu}^T{\bf w}$

$\lambda \geq 0$: trade-off parameter

The problem is highly non-convex due to the term $\sum^n_{i=1} g_i^2({\bf w})$.

#### Solution Algorithm

**1. Successive Convex Approximation (SCA)**

The idea behind SCA is that nonconvex problems are difficult to solve, so we solve a sequence of simpler problems (convex optimization):

1. Find a good enough approximation $\tilde{U}({\bf w},{\bf w}^{(k)})$ that approaches $U({\bf w})$ at ${\bf w}^{(k)}$.

2. Iterate

   $$
   \begin{align*}
   {\bf w}^{(k+1)} &= \text{arg } \underset{{\bf w}} {\text{min }}  \tilde{U}({\bf w};{\bf w}^{(k)})
   \\
   &\text{u.c. } \sum^n_{i=1} w_i = 1
   \end{align*}
   $$

3. Ensure $\tilde{U}({\bf w};{\bf w}^{(k)})$ satisfies:

   - $\tilde{U}({\bf w};{\bf w}^{(k)})$ is uniformly strongly convex and continuously differentiable.

   - $\nabla\tilde{U}({\bf w};{\bf w}^{(k)})$ is Lipschitz continuous on $W$.

   - $\nabla\tilde{U}({\bf w};{\bf w}^{(k)}) \big|_{{\bf w}={\bf w}^{(k)}} = \nabla U({\bf w}) \big|_{{\bf w}={\bf w}^{(k)}}$.

**Gradient Descent Method**

Choose($\alpha$ is the learning rate)

$$
\tilde{U}({\bf w},{\bf w}^{(k)})=U({\bf w}) + \nabla U({\bf w})^T ({\bf w}-{\bf w}^{(k)}) + \frac{1}{2\alpha^{(k)}} \lVert {\bf w}-{\bf w}^{(k)} \rVert^{2}_{2}
$$

Setting the derivative w.r.t. ${\bf w}$ to $0$ yields:

$$
{\bf w}^{(k+1)} = {\bf w}^{(k)} - \alpha^{(k)} \nabla U({\bf w}^{(k)})
$$

**Newton's Method**

Choose($\alpha$ is the learning rate)

$$
\tilde{U}({\bf w},{\bf w}^{(k)})=U({\bf w}) + \nabla U({\bf w})^T ({\bf w}-{\bf w}^{(k)}) + \frac{1}{2\alpha^{(k)}} ({\bf w}-{\bf w}^{(k)})^T \nabla^2 U({\bf w}^{(k)}) ({\bf w}-{\bf w}^{(k)})
$$

Setting the derivative w.r.t. ${\bf w}$ to $0$ yields:

$$
{\bf w}^{(k+1)} = {\bf w}^{(k)} - \alpha^{(k)} \left(\nabla^2 U({\bf w}^{(k)})\right)^{-1} \nabla U({\bf w}^{(k)})
$$

**2. Successive Convex Optimization(SCRIP)**

We've taken the nonconvex term $R({\bf w})$.

We want to convexified it by linearizing each $g_i({\bf w})$ within the square operation, and add the proximal term $\lVert {\bf w}-{\bf w}^{(k)} \rVert^{2}_{2}$ to ensure convergence.

The function $\tilde{U}({\bf w};{\bf w}^{(k)})$ is defined as follows($\tau>0$):

$$
\tilde{U}({\bf w};{\bf w}^{(k)}) = \sum^n_{i=1}[g_i({\bf w}^{(k)})+\nabla g_i({\bf w}^{(k)})^T({\bf w}-{\bf w}^{(k)})]^2+\frac{\tau}{2}\lVert {\bf w}-{\bf w}^{(k)} \rVert^2_2+\lambda F({\bf w})
$$

Let $P({\bf w};{\bf w}^{(k)}) = \sum^n_{i=1} [g_i({\bf w}^{(k)})+\nabla g_i({\bf w}^{(k)})^T({\bf w}-{\bf w}^{(k)})]^2$

**Idea**

Linear non-convex functions $g_i({\bf w})$ inside the square leads to quadratic convex $P({\bf w};{\bf w}^{(k)})$ that approximates $\mathcal{R}({\bf w})=\sum ^n_{i=1} g_i^2({\bf w})$

with $\nabla P({\bf w};{\bf w}^{(k)})|_{{\bf w}={\bf w}^{(k)}} = \nabla \mathcal{R}({\bf w})|_{{\bf w}={\bf w}^{(k)}}$

$P({\bf w};{\bf w}^{(k)})$ can be rewritten more compactly as:

$$
P({\bf w};{\bf w}^{(k)})=\lVert A^k({\bf w}-{\bf w}^{(k)})+g({\bf w}^{(k)}) \rVert^2
$$

where

$$
A^{(k)} = \left(\nabla g_1({\bf w}^{(k)}), \cdots, \nabla g_n({\bf w}^{(k)})\right)^T
$$  

$$
g({\bf w}^{(k)}) = \left(g_1({\bf w}^{(k)}), \cdots, g_n({\bf w}^{(k)})\right)^T
$$  

We can further expand $P({\bf w};{\bf w}^{(k)})$ as:

$$
P({\bf w};{\bf w}^{(k)}) = ({\bf w}-{\bf w}^{(k)})^T (A^{(k)})^T A^{(k)}({\bf w}-{\bf w}^{(k)})+g({\bf w}^{(k)})^T g({\bf w}^{(k)})+2g({\bf w}^{(k)})^T A^{(k)} ({\bf w}-{\bf w}^{(k)})
$$  

As a result, The quadratic program (QP) approximation problem at the k-th iteration is:

$$
\begin{align*}
\hat{\bf w}^{(k)} &= \text{arg } \underset{{\bf w}}{\text{min }} \tilde{U}({\bf w};{\bf w}^{(k)}) = \text{arg } \underset{{\bf w}}{\text{min }} \frac{1}{2}{\bf w}^T Q^{(k)} {\bf w} + {\bf w}^T q^{(k)} + \lambda F({\bf w})
\\
&\text{u.c. } \sum^n_{i=1} w_i = 1
\end{align*}
$$  

where

$$
Q^{(k)} = 2(A^{(k)})^T A^{(k)} + \tau I
$$  

$$
q^{(k)} = 2(A^{(k)})^T g({\bf w}^{(k)}) - Q^{(k)} {\bf w}^{(k)}
$$

and the iteration is given by:

$$
{\bf w}^{(k+1)} = {\bf w}^{(k)} + \gamma^{(k)} (\hat{\bf w}^{(k)} - {\bf w}^{(k)})
$$

where $\gamma^{(k)} \in (0,1]$ and $\gamma^{(k)} \to 0$

Under these assumptions, the problem is strongly convex and can be solved using existing efficient solvers.

**Algorithm**

- **Step 1**

    Choose an initial vector ${\bf w}^{(1)} \in \mathbb{R}^n$

- **Step 2**

    Use the following formula to update the vector ${\bf w}$

    $$
    \begin{align*}
    \hat{\bf w}^{(k)} &= \text{arg } \underset{{\bf w}}{\text{min }} \frac{1}{2}{\bf w}^T Q^{(k)} {\bf w} + {\bf w}^T q^{(k)} + \lambda F({\bf w})
    \\
    &\text{u.c. } \sum^n_{i=1} \hat{w}_i = 1
    \\
    {\bf w}^{(k+1)} &= {\bf w}^{(k)} + \gamma^{(k)} (\hat{\bf w}^{(k)} - {\bf w}^{(k)})
    \end{align*}
    $$

    where

    - $Q^{(k)} = 2(A^{(k)})^T A^{(k)} + \tau I$

    - $q^{(k)} = 2(A^{(k)})^T g({\bf w}^{(k)}) - Q^{(k)} {\bf w}^{(k)}$

    - $A^{(k)} = \left(\nabla g_1({\bf w}^{(k)}), \cdots, \nabla g_n({\bf w}^{(k)})\right)^T$  

    - $g({\bf w}^{(k)}) = \left(g_1({\bf w}^{(k)}), \cdots, g_n({\bf w}^{(k)})\right)^T$  

- **Step 3**

    Repeat step 2, until the solution ${\bf w}$ converge, or it's precise enough.

## Code Verification

In this case, we utilize the [cvxpy](https://github.com/cvxpy/cvxpy) package to solve for $\hat{{\bf w}}^{(k)}$

We specifically use the [cvxpy.MOSEK](https://docs.mosek.com/latest/pythonapi/intro_info.html) solver due to its superior efficiency.

Consider the function $g_i({\bf w})$ defined as follows:

$$
g_i({\bf w}) = \frac{w_i{(\Sigma{\bf w})}_i}{{{\bf w}^T\Sigma{\bf w}}}-\frac{1}{n} = \frac{{\bf w}^T M_i {\bf w}}{{{\bf w}^T\Sigma{\bf w}}}-\frac{1}{n}
$$

Here, $M_i$ is a sparse matrix where the i-th row is identical to that of the covariance matrix $\Sigma$.

The gradient of $g_i({\bf w})$ can then be computed as:

$$
\nabla g_i({\bf w}) =\frac{({\bf w}^T\Sigma {\bf w})(M_i+M_i^T){\bf w}-({\bf w}^TM_i{\bf w})(2\Sigma){\bf w}}{({{\bf w}^T\Sigma{\bf w}})^2}
$$

This formulation allows us to handle the general case in our optimization problem.

The gradient computation is crucial for the iterative update in the algorithm.

The sparse nature of the matrix $M_i$ can potentially lead to computational efficiency.

```python
def gradient(n,cov, w, i):
    Mi = np.zeros([n, n])
    Mi[i, :] = cov[i, :]
    return (((w.T @ cov @ w)*(Mi+Mi.T)@w-(w.T @ Mi @ w)*2*cov@w )/ (w.T @ cov @ w)**2 )

def g_i(w, cov, i, n):
    # Function to compute g_i(w)
    Mi = np.zeros([n, n])
    Mi[i, :] = cov[i, :]
    return ((w.T @ Mi @ w) / (w.T @ cov @ w) - 1/n)

def g(w, cov, n):
    # Function to compute the vector g(w)
    g = []
    for i in range(n):
        g.append(g_i(w, cov, i, n))
    return np.array(g).reshape([n, 1])

def A(w, cov, n):
    # Function to compute the matrix A(w)
    a = []
    for i in range(n):
        # gi = lambda w: g_i(w, cov, b, i, n)
        # a.append(gradient(gi, w))
        a.append(gradient(n,cov, w, i))
    a = np.array(a).reshape([n, n])
    return a

def big_Q(w, cov, n, tau):
    # Function to compute the matrix Q(w)
    e = np.eye(n)
    a = A(w, cov, n)
    return 2 * a.T @ a + tau * e

def small_q(w, cov, n, tau):
    # Function to compute the vector q(w)
    a = A(w, cov, n)
    return 2 * a.T @ g(w, cov, n) - big_Q(w, cov, n, tau) @ w

def SCRIP_computation(x0,tau,cov,n,lambda_,mu,max_err=1e-6,times=1000):
    # Function to perform SCRIP computation

    err = 520
    gamma=0.9
    # iterate time
    t=0
    x=[x0]
    k=0
    w = cp.Variable([n, 1])

    while err >= max_err and t < times:
        # optimize \hat{w}
        obj=cp.Minimize((1 / 2) * cp.quad_form(w, big_Q(x[k],cov,n,tau)) + (w.T @ small_q(x[k], cov, n, tau)))
        con=[cp.sum(w) == 1, w >= 0]
        prob = cp.Problem(obj,con)
        prob.solve() # cp.MOSEK  cp.CVXOPT

        x.append(x[k]+gamma*(w.value-x[k]))

        gamma=gamma*(1-1e-7*gamma)

        t += 1
        k += 1
        err = np.linalg.norm(x[k] - x[k-1], ord=np.inf)

    return x

def SCRIP(n, mu, cov,lambda_=0):
    # Function to perform SCRIP optimization for a portfolio

    # parameters
    tau= 0.05*np.trace(cov)/(2*n)
    # initial guess
    x0 = np.ones([n, 1]) / n

    ans = SCRIP_computation(x0,tau,cov,n,lambda_,mu)
    return ans
```
